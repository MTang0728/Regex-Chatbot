```python
import numpy as np
import pandas as pd

import re
import random
```

```python
filepath = './input.txt'
sample_prompts = []
with open(filepath) as f:
    data = f.readlines()
    for line in data:
        sample_prompts.append(line.strip())
```

```python
sample_prompts
```

```python
modeldict = {
    "(.*)?how('re| a?re?) (yo)?u(.*)?":[
        "I'm good! How can I help you?"
    ],
    '.*(he+l+o+|h+i+).*':[
        'Hello! How can I help you?'
    ],
    '(w.*)?s+.*u+p+(.*)?':[
        'Not much, you?',
        "I'm ok, how about you?",
        "What's up with you?"
    ],
    '.*wh(at|o).*(are|r) (you|u).*':[
        "I'm the almighty!",
        'Just your ordinary chatbot.'
    ],
    '(.*)?you do(.*)?': [
        "What do I do? That is a thinker.",
        'What do you think I do?',
        'I answer questions!'
    ],
    '((?=.*who)|((?=.*what)(?=.*name)))(?=.*coach)(?=.*duke)?(?=.*basketball)?(?=.*k)':[
        "Mike Krzyzewski.",
        'I think you are talking about Coach Mike Krzyzewski.',
        'Mike Krzyzewski - head coach of Duke basketball team.'
    ],
    '.*long.*(coach k|(mike|michael) kr?zyz?ew?sk[iy]).*coached.*':[
        'Over 40 years.',
        'A really long time.'
    ],
    '.*where.*duke.*basketball.*':[
        'Cameron Indoor Stadium.'
    ],
    '.*(what|wut).*duke.*mascou?t.*':[
        'Blue Devil.',
        'Something Blue, something devilish.'
    ],
    '.*what.*(k|kr?zyz?ew?sk[iy])(-*)?ville?.*':[
        "I guess it's a fictious place?",
        'A basketball cult phenomenon at Duke University.'
    ]
}
```

```python
default = [
    'Hmmmm, let me think about that',
    'Clearly you are not very sophisticated'
]
```

```python
matched=0

for prompt in sample_prompts:
    for pattern in modeldict:
        if re.search(pattern,prompt.lower()):
            matched = 1
            ans=random.choice(modeldict[pattern])
            pass
        pass
    if matched==0:
        ans=random.choice(default)
        pass
    print(f'prompt: {prompt}\nanswer: {ans}',sep='\n')
    print('\n')
    pass
```
